```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read file and turn letters in lower cases
Eng = open("Alice_eng.txt")
eng = Eng.read().lower()

Ger = open("Alice_ger.txt")
ger = Ger.read().lower()

Ita = open("Alice_ita.txt")
ita = Ita.read().lower()

Fin = open("Alice_fin.txt")
fin = Fin.read().lower()

Esp = open("Alice_esp.txt")
esp = Esp.read().lower()
```

```python
# define counting function
def count_letters(file):
    dictionary = {}
    for letter in "abcdefghijklmnopqrstuvwxyz":
        dictionary[letter] = 0
    for letter in file:
        if letter in "abcdefghijklmnopqrstuvwxyz":
            dictionary[letter] = dictionary[letter]+1
    return dictionary 
```

```python
dic_eng = count_letters(eng)
dic_ger = count_letters(ger)
dic_ita = count_letters(ita)
dic_fin = count_letters(fin)
dic_esp = count_letters(esp)
```

```python
eng_df = pd.DataFrame.from_dict(dic_eng, orient='index', columns=['ENG'])
eng_df = pd.DataFrame.from_dict(dic_eng, orient='index', columns=['ENG'])
ger_df = pd.DataFrame.from_dict(dic_ger, orient='index', columns=['GER'])
ita_df = pd.DataFrame.from_dict(dic_ita, orient='index', columns=['ITA'])
fin_df = pd.DataFrame.from_dict(dic_fin, orient='index', columns=['FIN'])
esp_df = pd.DataFrame.from_dict(dic_esp, orient='index', columns=['ESP'])
```

```python
count_def = pd.concat([eng_df, ger_df, ita_df, fin_df, esp_df],axis=1)
```

```python
count_def
```

```python
eng_pro = eng_df/sum(dic_eng.values())
ger_pro = ger_df/sum(dic_ger.values())
ita_pro = ita_df/sum(dic_ita.values())
fin_pro = fin_df/sum(dic_fin.values())
esp_pro = esp_df/sum(dic_esp.values())
```

```python
pro_def = pd.concat([eng_pro, ger_pro, ita_pro, fin_pro, esp_pro],axis=1)
```

```python
pro_def
```

### (a) Compute all pairwise variational distance

```python
def compute_distance(a_pro,b_pro): # let a_pro, b_pro indicate the eng_pro etc. as above pdframe.
    distance = 0.5 * sum(abs(a_pro.values - b_pro.values))
    return distance
```

```python
eng_ger = compute_distance(eng_pro,ger_pro)
eng_ita = compute_distance(eng_pro,ita_pro)
eng_fin = compute_distance(eng_pro,fin_pro)
eng_esp = compute_distance(eng_pro,esp_pro)
ger_ita = compute_distance(ger_pro,ita_pro)
ger_fin = compute_distance(ger_pro,fin_pro)
ger_esp = compute_distance(ger_pro,esp_pro)
ita_fin = compute_distance(ita_pro,fin_pro)
ita_esp = compute_distance(ita_pro,esp_pro)
fin_esp = compute_distance(fin_pro,esp_pro)
```

```python
distance = [eng_ger, eng_ita, eng_fin, eng_esp, ger_ita, ger_fin, ger_esp, ita_fin, ita_esp, fin_esp]
```

```python
distance_df = pd.DataFrame(distance,columns=['distance between pairs'])
distance_df.index = ['eng_ger','eng_ita','eng_fin','eng_esp',
                     'ger_ita','ger_fin','ger_esp',
                     'ita_fin','ita_esp',
                     'fin_esp']
```

```python
distance_df 
```

### (b) Compute the five collision probabilities

```python
def compute_collision(a_pro):
    collision = sum(np.power(a_pro.values,2))
    return collision
```

```python
eng_col = compute_collision(eng_pro)
ger_col = compute_collision(ger_pro)
ita_col = compute_collision(ita_pro)
fin_col = compute_collision(fin_pro)
esp_col = compute_collision(esp_pro)
```

```python
collision = [eng_col, ger_col, ita_col, fin_col, esp_col]
```

```python
collision_df = pd.DataFrame(collision,columns=['collision'])
collision_df.index = ['eng_col','ger_col','ita_col','fin_col','esp_col']
```
